```python
import pandas as pd
import numpy as np
```

```python
#Importing Data
```

```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
```

```python
rating_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
ratings = pd.read_csv(rating_url, sep='\t', low_memory=False)
```

```python
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

```python
#AKAs
```

```python
akas.info()
akas.head()
```

```python
akas = akas.replace({'\\N':np.nan})
```

```python
us = akas['region'] == 'US'
akas = akas[us]
akas
```

```python
#Basics
```

```python
basics.info()
basics.head()
```

```python
basics = basics.replace({'\\N':np.nan})
```

```python
basics = basics.dropna(subset=["runtimeMinutes","genres"])
basics
```

```python
Movie = basics['titleType'] <= 'movie'
basics = basics[Movie]
basics
```

```python
twothou = basics['startYear'] > '1999'
basics = basics[twothou]
basics
```

```python
#As we're only looking at movies, we don't need to look at endYear since it shows the year TV shows in the data ended
basics = basics.drop(columns=['endYear'])
basics.head()
```

```python
docs =basics['genres'].str.contains('documentary',case=False)
basics = basics[~docs]
basics
```

```python
keepers = basics['tconst'].isin(akas['titleId'])
keepers
```

```python
basics = basics[keepers]
basics.info()
basics
```
